```
#Funcion para hallar la frecuencia de una seal 
def fft_plot(t, y, wmin=None, wmax=None):
    """
    Esta funcin recibe una seal (con su vector de tiempo) y grfica 
    su representacin en el tiempo y en la frecuencia.
    
    t:  vector de tiempo
    y: seal
    wmin: frecuencia mnima en la grfica
    wmax: frecuencia mxima en la grfica
    
    """
    # La frequencia de muestreo la extraigo del vector del tiempo
    sample_rate = t[1] - t[0]
    
    Y = np.abs(fftshift(fft(y)))  # Valor absoluto de Fourier
    Y = Y / Y.shape[0]  # Normalizacin
    W = fftshift(fftfreq(Y.shape[0], sample_rate))  # Vector de frecuencias
    
    #  Inicializa la figura con un tamao y densidad de pxeles cmoda
    plt.figure(figsize=(16, 5), dpi=90)
    
    #  Primer plot
    plt.subplot(121)  # 1 fila, 2 columnas, espacio 1
    plt.plot(t, y, color='C0')
    plt.grid(True)
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    
    plt.subplot(122)  # 1 fila, 2 columnas, espacio 2
    plt.plot(W, Y, color='C1')  # grafica el contorno
    plt.fill_between(W, Y, 0, color='C1', alpha=0.3)  # grafica el relleno con una transparencia
    if wmin and wmax:
        plt.xlim(wmin, wmax)
    plt.ylim(0)
    plt.grid(True)
    plt.xlabel('Frequency [Hz]')
    plt.ylabel('Amplitude')
```

```
def normalization(signal):
    """
    Normaliza la seal, dejndola nicamente positiva y variando de 0 a 1
    
    signal: seal
    """
    signal_ = signal.copy()
    signal_ -= signal_.min()  # Se desplaza para que nicamente sea positiva
    signal_ /= signal_.max()  # Se divide por el mximo, para que el dominio sea de 0 a 1
    
    return signal_
l=normalization(output_c)#Se normaliza la salida del circuito
fft_plot(t,l,-5,5)
```

##AM##
**Modulacin de doble banda lateral con portadora suprimida (DSB-SC)**

$$y(t)=a(t)cos(w_ct)$$

En donde

$$a(t)\sim  kx(t)+k_d \hspace{1cm} k,k_d=cte$$

Para esta modulacin se asume $k_d=0$ por lo que se tiene

$$y(t)=x(t)cos(w_ct)$$

Cuya densidad espectral esta dada por

$$Y(w)=F(y(t))=\frac{X(w+w_c)+X(w-w_c)}{2}$$

```
u = 1

DSBSC = l*portadora*u

fft_plot(t, DSBSC, -10,10)
```

**Modulacin de doble banda lateral con gran portadora (DSB-LC)**

Se obtienen, si se asume una constante de desplazamiento $k_d$ diferente de cero, por lo que la seal modulada sera

$$y(t)=x(t)cos(w_ct)+k_dcos(w_ct)$$

donde, la correspondiente densidad espectral estara dada por

$$Y(w)=\frac{1}{2}X(w+w_c)+\frac{1}{2}X(w-w_c)+k_d \pi \frac{1}{2}\delta(w+w_c)+k_d \pi \frac{1}{2}\delta(w-w_c)$$

```
u = 1
kd=0.5
DSBLC = u*(l*portadora+kd*portadora)

fft_plot(t, DSBLC, -10,10)
```

**Modulacin de banda lateral nica (SSB)**
Se basa transmitir el mensaje utilizando solo una de las bandas laterales, se toma una seal DSB de la cual se suprimeuna de las bandas laterales con un filtro.

$$y(t)=x(t)cos(w_ct) \mp \hat x(t)sin(w_ct)$$

Donde la suma representa la banda lateral inferior

$$y(t)=x(t)cos(w_ct)+\hat x(t)sin(w_ct)$$

mientras que la diferencia, representa la banda lateral superior

$$y(t)=x(t)cos(w_ct) - \hat x(t)sin(w_ct)$$

La densidad espectral sera

$$F(y(t))=F(x(t)cos(w_ct)) \mp F(\hat x(t)sin(w_ct))$$
$$Y(w)= \frac{1}{2}X(w-w_c)+\frac{1}{2}X(w+w_c)-(\frac{1}{2j}\hat X(w-w_c)-\frac{1}{2j}\hat X(w+w_c))$$

En donde 

$$\hat X(w \mp w_c)=-jsgn(w \mp w_c)X(w \mp w_c)$$

por lo que 

$$Y(w)= \frac{1}{2}X(w-w_c)(1+sgn(w - w_c))+\frac{1}{2}X(w+w_c)(1-sgn(w + w_c))$$

sabiedo que la funcion signo es

$$sgn(w-wc)=\left\{\begin{matrix}
1 ; w>w_c\\ 
0 ; w<w_c
\end{matrix}\right.$$

En este caso 

$$1 \mp sgn(w-wc)=\left\{\begin{matrix}
2 ; w> \mp w_c\\ 
0 ; w< \mp w_c
\end{matrix}\right.$$

y por lo tanto

$$Y(w)=\left\{\begin{matrix}
0 \hspace{3cm}   ; |w|>w_c\\ 
X(w+w_c) \hspace{1cm} ; w<-w_c\\
X(w-w_c) \hspace{1cm} ; w>w_c
\end{matrix}\right.$$

```
from scipy.signal import hilbert

segport= np.cos(2*np.pi*fp*(t+np.pi))

#banda lateral superior 
SSBS= DSBSC + u*hilbert(l)*segport
fft_plot(t, SSBS, -10, 10)

#banda lateral inferior 
SSBI= DSBSC - u*hilbert(l)*segport
fft_plot(t, SSBI, -10, 10)
```

**Modulacin por amplitud de pulso (PAM)**

para este caso la seal portadora consiste en un tren de pulsos y la modulacin se basa en variar algn parmetro de la seal portadora.

La modulacin por amplitud de pulso consiste en la multiplicacion de la seal analogica contnua del mensaje, por un tren de pulsos cuadrados constantes en amplitud, de donde se modula la amplitud de acuerdo al mensaje a enviar

$$y_{PAM}(t)=\sum_K x(KT)rect_\tau (t-KT)$$

donde T es el periodo de muestreo, $x(KT)$ es el valor muestreado de $x(t)$ obteido en el tiempo $t=KT$ y la sumatoria de las funciones $rect_\tau (t)$ representan el tren de pulsos de amplitud $1$ y duracion $\tau$.

Tambin se puede encontral la seal modulada realizando la convolucion entre el mensaje muestreado instantaneamente $x(t)$ y el pulso rectangular $rect_\tau (t)$, donde, el mensaje muestread de forma instantanea esta dado por

$$x_\delta (t)=\sum_K x(KT) \delta (t-KT)$$

entonces

$$y_{PAM}(t)=x_\delta (t)*rect_\tau (t)=\sum_K x(KT)rect_\tau (t-KT)$$

La densidad espectral se encuentra aplicando la propiedad de la transformada de Fourier para la convolucin

$$F(y_{PAM}(t))=F(x_\delta (t)*rect_\tau (t))=X_\delta (w) F(rect_\tau (t))$$
$$Y_{PAM}(w)= \frac{\tau}{T} \sum_K X(w-Kw_o) sinc(w \frac{\tau}{2}) $$

**Definimos tren de pulsos como portadora**

```
def pulse_train(t, pulses=10):
    """
    Define un tren de pulsos.
    
    t: vector de tiempo
    pulses: nmero de pulsos
    
    """
    width = (t.max()-t.min())/(pulses)
    
    return t, np.array(t%width*2<(width), dtype=np.int)

pulses = 50
t, tren = pulse_train(t, pulses)

fft_plot(t, tren)
```

```
#muestreo natural
PAMN = tren * l
fft_plot(t, PAMN)

#muestreo con pulsos de amplitud plana
PAMR = np.array([[pulse.mean()]*pulse.shape[0] for pulse in PAMN.reshape(-1, PAMN.shape[0]//(pulses*2))]).reshape(-1, )
fft_plot(t, PAMR)
```

## Modulacion por ancho de pulso (PWM)##
La modulacin por ancho de pulso (PWM) consiste en variar la duracion de los pulsos de la portadora (Tren de pulsos), segn la forma que tenga la seal moduladora.

La generacin tanto de la modulacion por ancho de pulso, como la modulacin por posicin de pulsos posteriormente mencionada consiste en la comparacin de la seal moduladora con una seal diente de sierra, en donde

La modulacin por ancho de pulsos, compara la distancia entre el comienzo de periodo de la seal diente de cierra hasta la interseccin que tiene esta con la seal moduladora, representando este ancho como el ancho de cada pulso, es decir, entre mayor sea la amplitud de la seal moduladora, mayor ancho tendra el pulso de la portadora.

```
lp = l - min(l)
diente_sierra = (np.max(lp))*(signal.sawtooth( 2* np.pi*t)+1)/2

fft_plot(t, diente_sierra, -10, 10)
```

```
comparation = diente_sierra - lp
comparation = np.abs(comparation)
fig,ax = plt.subplots((1),figsize=(16, 9), dpi=90)
lista = []
PWM = np.zeros_like(lp)
PPM = np.zeros_like(lp)
for i in range(0,10):
  index = 200*i
  index2= 200*(i+1)
  lista.append(np.argmin(comparation[index+1:index2]))
  PWM[index:lista[i]+(200*i)] = np.max(lp)
  PPM[lista[i]+(200*i):lista[i]+(200*i) + 20] = np.max(lp)

ax.title.set_text("Modulacin PWM")
ax.plot(t, PWM, '-', label="Seal modulada")
ax.plot(t, lp, '--', label="Seal original")
# plt.plot(t, abs(lp), '--')
ax.set_xlabel('Amplitude')
ax.set_ylabel('Time')
ax.grid(True)
ax.legend()

plt.plot()
```
