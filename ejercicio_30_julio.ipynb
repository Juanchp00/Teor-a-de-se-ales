```python
import numpy as np
import random
señal=np.zeros(64)
for i in range (64):
  if (i<=32):
    señal[i]=1
  else:
    señal[i]=-1

for i in range(3):
  random.shuffle(señal)
  
señal
```

**señal**

```python
import numpy as np
from matplotlib import pyplot as plt
s1=1
s2=-1

symbols= int(2**12)
cantidad=128

señal= np.random.choice([s1,s2], size=cantidad, p=[0.5, 0.5])
señal=señal.astype('float')

duracion=int(symbols/cantidad)
señal=np.concatenate([[i]*duracion for i in señal])
señal.shape
plt.plot(señal)
```

**Ruido blanco gaussiano**


```python
t = np.linspace(0, 10, 4096)
noise = np.random.normal(0,0.1,len(t))
plt.plot(noise)
```

**señal con ruido**

```python
syn=señal+noise
plt.plot(syn)
```
